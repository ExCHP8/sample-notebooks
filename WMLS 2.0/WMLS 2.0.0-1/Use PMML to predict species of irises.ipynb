## 2. Save a PMML model

In this section, you will use the Python API client to create an online deployment and score a new data record against the PMML model.

For more information, please refer to the <a href="https://wml-api-pyclient-dev-v4.mybluemix.net/" target="_blank" rel="noopener no referrer">V4 Python client documentation</a>.


Use `wget` to download the sample PMML model, `iris_chaid.xml` from the <a href="https://github.com/pmservice/wml-sample-models" target="_blank" rel="noopener no referrer">GitHub repository</a>.  

<div class="alert-block alert-info"><br>
&nbsp;&nbsp;&nbsp;&nbsp;You may need to install the <tt>wget</tt> package. To install the <tt>wget</tt> package, run the following command.<br><br></div>

```python
!pip install -q --upgrade wget
```

```python
# Download the sample PMML model and the 'iris_chaid.xml' file from the GitHub repo.
from os import mkdir
from os.path import isdir, isfile, join
from wget import download

sample_dir = 'pmml_sample_model'
if not isdir(sample_dir):
    mkdir(sample_dir)

filename = join(sample_dir, 'iris_chaid.xml')
if not isfile(filename):
    filename = download(
        'https://github.com/pmservice/wml-sample-models/raw/master/pmml/' +
        'iris-species/model/iris_chaid.xml',
        out=sample_dir)
```

Store the downloaded file as *CHAID PMML model for Iris data*, and then list all the saved models. First, you need to create the model metadata.


```python
# Model metadata
props_pmml = {
    client.repository.ModelMetaNames.NAME: 'CHAID PMML model for Iris data',
    client.repository.ModelMetaNames.RUNTIME_UID: 'pmml_4.2.1', 
    client.repository.ModelMetaNames.TYPE: 'pmml_4.2.1'
}
```

You can extract the model UID from the saved model details.

```python
# Create the model artifact.
model_artifact = client.repository.store_model(filename, meta_props=props_pmml)
model_uid = client.repository.get_model_uid(model_artifact)
print('Model UID = {}'.format(model_uid))
```

Obtain the saved model metadata from WML using the model UID.

```python
from json import dumps

# Details about the model.
model_details = client.repository.get_details(model_uid)
print(dumps(model_details, indent=4))
```

You can list all stored models using the `list_models` method.

```python
# Display a list of all the models.
client.repository.list_models()
```

<div class="alert-block alert-info"><br>
&nbsp;&nbsp;&nbsp;&nbsp;From the list of stored models, you can see that model is successfully saved. <br>

&nbsp;&nbsp;&nbsp;&nbsp;With Watson Studio Desktop credentials, this means you have saved the model in your project.<br>
&nbsp;&nbsp;&nbsp;&nbsp;You can see the saved model in your project UI by clicking on your project name in the breadcrumb at the top of the application. <br>

## 3. Create an online deployment (WML Server only) <a id="deploy"></a>

Now, create an online deployment, *Iris species prediction*, for the stored model, and list all the online deployments in the deployment space. You will create deployment metadata and use the model UID obtained in the previous section.


```python
# Deployment metadata.
deploy_meta = {
    client.deployments.ConfigurationMetaNames.NAME: 'Iris species prediction',
    client.deployments.ConfigurationMetaNames.ONLINE: {}
}
```

```python
# Create an online deployment.
deployment_details = client.deployments.create(model_uid,
                                               meta_props=deploy_meta)
```

```python
# List the deployments.
client.deployments.list()
```

<div class="alert-block alert-info"><br>
&nbsp;&nbsp;&nbsp;&nbsp;From the list of deployed models in the deployment space, you can see that model was successfully created.<br><br></div>

Now, you can check details of your deployments.

```python
# Deployment UID.
deployment_uid = client.deployments.get_uid(deployment_details)
print('Deployment UID = {}'.format(deployment_uid))
```

You can check the deployment details by running the following cell.

```python
# Deployment details.
print(dumps(client.deployments.get_details(deployment_uid), indent=4))
```

### 3.1 Score a test data record<a id="score"></a>

Now, you can predict the species of iris flowers. Create a scoring payload.

```python
# Predict the species of an iris flower. Prepare scoring payload.
job_payload = {
    client.deployments.ScoringMetaNames.INPUT_DATA: [{
        'fields':
        ['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width'],
        'values': [[5.1, 3.5, 1.4, 0.2]]
    }]
}
print(dumps(job_payload, indent=4))
```

```python
# Predict the species of an iris flower.
predictions = client.deployments.score(deployment_uid, job_payload)
print(dumps(predictions, indent=4))
```

As we can see from the prediction, the species of the iris flower is setosa.

### 3.2 Delete the deployment and model<a id='delete'></a>

Use the following method to delete the deployment.

```python
client.deployments.delete(deployment_uid)
```

```python
client.deployments.list()
```

You can delete the model as well by running the following cell.

```python
client.repository.delete(model_uid)
```

You can check that your model was deleted by generating a list of your saved models.

```python
client.repository.list_models()
```
